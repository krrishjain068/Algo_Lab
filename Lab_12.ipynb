# MATRIX CHAIN MULTIPLICATION Dynamic Programming with table filling.
Problem Description:
Find minimum number of scalar multiplications needed to multiply a chain of matrices..

#Theory
Cost of multiplying two matrices:
If A is p×q and B is q×r, cost = p × q × r

Dynamic Programming idea:
Break problem into smaller subproblems.
Store results in a table to avoid recomputation.

m[i][j] = minimum cost to multiply Ai to Aj

## Algorithm
1. Take array p[] of size n
2. Create table m[n][n]
3. Set m[i][i] = 0
4. Fill table diagonally
5. For each chain length L:
    Compute m[i][j]
    Try all possible k between i and j
6. Choose minimum cost
7. Return m[1][n-1]

```python
import sys

def matrix_chain_multiplication(p):
    n = len(p)

    # Create DP table
    m = [[0 for _ in range(n)] for _ in range(n)]

    # L is chain length
    for L in range(2, n):
        for i in range(1, n - L + 1):
            j = i + L - 1
            m[i][j] = sys.maxsize

            for k in range(i, j):
                cost = (m[i][k] +
                        m[k+1][j] +
                        p[i-1] * p[k] * p[j])

                if cost < m[i][j]:
                    m[i][j] = cost

    return m

```

```python
# Input dimensions
p = [10, 30, 5, 60]

table = matrix_chain_multiplication(p)

print("Minimum multiplication cost:", table[1][len(p)-1])

print("\nDP Table:")
for row in table[1:]:
    print(row[1:])
```

# Exercise
1. Given p = [5, 10, 3, 12, 5, 50, 6]
2. Compute minimum cost
3. Print optimal parenthesization
4. Compare number of multiplications saved

### Solution to Exercise
To find the optimal parenthesization, we need to modify the DP algorithm to store the split points `k` that yield the minimum cost for each subproblem. We will use a separate table, `s`, for this purpose.

```python

p_exercise = [5, 10, 3, 12, 5, 50, 6]
```

```python
import sys

def matrix_chain_multiplication_with_parenthesization(p):
    n = len(p)

    m = [[0 for _ in range(n)] for _ in range(n)]

    s = [[0 for _ in range(n)] for _ in range(n)]

    # L is chain length
    for L in range(2, n):
        for i in range(1, n - L + 1):
            j = i + L - 1
            m[i][j] = sys.maxsize

            for k in range(i, j):
                cost = (m[i][k] +
                        m[k+1][j] +
                        p[i-1] * p[k] * p[j])

                if cost < m[i][j]:
                    m[i][j] = cost
                    s[i][j] = k
    return m, s

def print_optimal_parenthesization(s, i, j):
    if i == j:
        return f"A{i}"
    else:
        left = print_optimal_parenthesization(s, i, s[i][j])
        right = print_optimal_parenthesization(s, s[i][j] + 1, j)
        return f"({left}{right})"
```

```python
# Compute the minimum cost and optimal split points for the exercise input
min_cost_table, split_points_table = matrix_chain_multiplication_with_parenthesization(p_exercise)

n_exercise = len(p_exercise)
minimum_cost_exercise = min_cost_table[1][n_exercise - 1]

print(f"For p = {p_exercise}:")
print(f"Minimum multiplication cost: {minimum_cost_exercise}")

print("\nOptimal Parenthesization:")
optimal_paren_str = print_optimal_parenthesization(split_points_table, 1, n_exercise - 1)
print(optimal_paren_str)

print("\nDP Cost Table:")
for row in min_cost_table[1:]:
    print(row[1:])

print("\nSplit Points Table (s):")
for row in split_points_table[1:]:
    print(row[1:])
```

### Comparison of Multiplications Saved

Comparing the number of multiplications saved requires understanding the naive approach vs. the dynamic programming approach.

A naive approach would involve trying all possible parenthesizations, which can be computationally very expensive (exponential complexity). The number of ways to parenthesize a product of `n` matrices is given by the Catalan number `C(n-1)`.

For `n_exercise = 7` matrices (since `p` has 7 elements, there are 6 matrices: A1 (5x10), A2 (10x3), ..., A6 (50x6)), the number of possible parenthesizations would be `C(6) = 4862`.

Calculating the cost for each of these parenthesizations and finding the minimum would involve many more operations than the DP approach, which has a time complexity of O(n^3).

However, directly comparing the 'number of multiplications saved' in terms of scalar operations is not straightforward without defining a specific naive strategy to compare against. The benefit of DP is avoiding redundant calculations and ensuring optimality efficiently, rather than a direct count of 'saved' scalar multiplications from a specific non-optimal path.
